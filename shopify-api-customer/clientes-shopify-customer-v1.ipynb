```python
#importar librerías y módulos
import requests
import os
from dotenv import load_dotenv
```

```python
#cargar las variables del archivo .env a jupyter
load_dotenv()
```

```python
#asignar variables a las credenciales que están en la carpeta .env
shop_name = os.getenv("SHOP_NAME")
api_key = os.getenv("SHOPIFY_API_TOKEN")
```

```python
#La parte final varía según la data que necesitamos - este link es estandar
url_base = f"https://{shop_name}.myshopify.com/admin/api/2025-07/customers.json"
```

```python
headers = {
    "X-Shopify-Access-Token": api_key,
    "Content-Type": "application/json"
        }
```

```python
respuesta = requests.get(url_base,headers=headers)
```

```python
data = respuesta.json()
```

```python
clientes = []

for c in data["customers"]:
    cliente_info = {
        "nombre": c["first_name"],
        "email": c["email"],
        "telefono": c["default_address"]["phone"],
    }
    clientes.append(cliente_info)
```
