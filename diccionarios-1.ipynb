```python
#Diccionario es una estructura de datos que guarda información en pares clave–valor

persona = {
    "nombre": "Alvaro",
    "edad": 30,
    "email": "alvaro@gmail.com"
}

persona['email']
```

```python
#El valor de una clave es una lista
persona = {
    "nombre": "Alvaro",
    "edad": 30,
    "email": ["alvaro@gmail.com","alvaro@hotmail.com","alvaro@empresa.com"]
}

persona['email'][2]
```

```python
#El valor de una clave es diccionario
persona = {
    "nombre": "Alvaro",
    "edad": 30,
    "contacto": {"email_1": "alvaro@gmail.com",
                 "email_2": "alvaro@hotmail.com",
                 "email_3":"alvaro@empresa.com"}
}

persona['contacto']['email_3']
```

```python
#El valor de la clave es una lista de diccionarios
persona = {
    "nombre": "Alvaro",
    "edad": 30,
    "contacto": [
        {"email": "alvaro@gmail.com","email_2": "alvaro@hotmail.com","email_3":"alvaro@empresa.com"},
        {"facebook": "alvaroenfacebook","instagram": "alvaroeninstagram","tiktok":"alvaroentiktok"}
    ]
}

persona['contacto'][1]['tiktok']
```
