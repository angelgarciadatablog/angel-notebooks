# Uso de FOR


```python

    # Código a ejecutar en cada vuelta
```

````python
#recorrer una lista de números

lista = [1,2,3,4,5]

for variable in lista:
    print(variable)
````

````python
#recorrer una lista con condicionales

lista = [3,7,2,8,4]

for variable in lista:
    if variable % 2 ==0:
        print(variable)
````

````python
#recorrer una lista con condicionales

lista = [3,7,2,8,4]

for variable in lista:
    if variable % 2 == 0:
        print(variable,"es par")
    else:
        print(variable,"es impar")
````

````python
#sumar todos los elementos de una lista


lista = [1, 2, 3, 4, 5]
total = 0

for variable in lista:
    total = total + variable  ##acumulando la suma de la lista

print (total) #imprime el acumulado
````

````python
#sumar todos los elementos de una lista

lista = [1, 2, 3, 4, 5]
total = 0

for variable in lista:
    total = total + variable #acumula la suma en cada iteración
    print (total) #imprime cada iteración
````

````python
#recorrer una lista de diccionario

personas = [
    {"nombre": "Ana", "edad": 25},
    {"nombre": "Luis", "edad": 30}
]

for variable in personas:
    print(variable)
````

````python
#recorrer una lista de diccionario

personas = [
    {"nombre": "Ana", "edad": 25},
    {"nombre": "Luis", "edad": 30}
]

for variable in personas:
    print(variable["nombre"])
````

## Recuento de palabras repetidas

````python
texto = "Python es genial y Python es divertido"

# Convertimos todo a minúsculas y separamos por palabras
palabras = texto.lower().split()

print(palabras)
````

````python
texto = "Python es genial y Python es divertido"
palabras = texto.lower().split()
recuento_palabras = {}

for variable in palabras:
    if variable in recuento_palabras:
        recuento_palabras[variable] = recuento_palabras[variable] + 1
    else:
        recuento_palabras[variable] = 1

print (recuento_palabras)      
````
