## Todos estos métodos pertenecen al tipo de dato str de Python y no requieren importar ninguna librería adicional.

```python
#El método .split() se aplica a un string (cadena de texto) y lo convierte en una lista de subcadenas, 
#separadas por un separador (por defecto, el espacio).

"Hola mundo".split()  # ['Hola', 'mundo']
```

```python
# strip() Elimina espacios (u otros caracteres) al inicio y final.
"  hola  ".strip()
```

```python


"Hola".lower()      
"Hola".upper()      
"hola mundo".capitalize() 
```

```python
# replace() Reemplaza todas las apariciones de un texto por otro.
"hola mundo".replace("mundo", "Python")
```

```python
# count() Cuenta cuántas veces aparece un texto.
"banana".count("a")
```

```python
#endswith() Verifica si la cadena empieza o termina con cierto texto.
"archivo.txt".endswith(".txt")
```

```python


"123".isdigit()     # True
"abc".isalpha()     # True
"abc123".isalnum()  # True
```
