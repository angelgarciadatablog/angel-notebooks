## Recuento de palabras

### prueba

```python
#Envuelvo el texto que voy a analizar en una variable
texto = "Python es genial y Python es divertido"

#Convertirmos todo a minúscula, lo separamos por palabras y lo metemos a una lista de palabras
lista_palabras = texto.lower().split()


#Creamos un diccionario vacio, lo llamaremos libreta_conteo
libreta_conteo = {}

#En este caso a cada palabra de la lista se le está asignando un 1 
for palabra in lista_palabras:
    libreta_conteo[palabra] = 1
    

print(libreta_conteo)
```

### real

```python
# Envuelvo el texto que voy a analizar en una variable
texto = "Python es genial y Python es divertido"

# Convertimos todo a minúscula, lo separamos por palabras y lo metemos a una lista de palabras
lista_palabras = texto.lower().split()

# Creamos un diccionario vacío, lo llamaremos libreta_conteo
libreta_conteo = {}

# En este caso se ha agregado un condicional que dice:
# Si la palabra (elemento de la lista) ya está en el diccionario,
# entonces se suma 1 al valor de esa clave (la palabra).
# Si no está, se agrega por primera vez con el valor 1.
for palabra in lista_palabras:
    if palabra in libreta_conteo:
        libreta_conteo[palabra] = libreta_conteo[palabra] + 1
    else: 
        libreta_conteo[palabra] = 1

# Resultado: nos muestra cuántas veces aparece cada palabra en el texto.
print(libreta_conteo)
```

# Convertimos a función

```python
#la primera variable que se necesita de entrada, se le rodea en una función (cualquier nombre)
def contar_palabras(texto):
    
    # Convertimos todo a minúscula, lo separamos por palabras y lo metemos a una lista de palabras
    lista_palabras = texto.lower().split()
    
    # Creamos un diccionario vacío, lo llamaremos libreta_conteo
    libreta_conteo = {}
    
    # En este caso se ha agregado un condicional que dice:
    # Si la palabra (elemento de la lista) ya está en el diccionario,
    # entonces se suma 1 al valor de esa clave (la palabra).
    # Si no está, se agrega por primera vez con el valor 1.
    for palabra in lista_palabras:
        if palabra in libreta_conteo:
            libreta_conteo[palabra] = libreta_conteo[palabra] + 1
        else: 
            libreta_conteo[palabra] = 1


    #se cambia el print por return (debe estar en una sola línea)
    return libreta_conteo
```

```python
#Probamos la función:

frase = "Soy una serpiente Que anda por el bosque Buscando una parte de su cola ¿Quiere ser usted una parte de mi cola?" 

prueba = contar_palabras(frase)

print(prueba)
```
